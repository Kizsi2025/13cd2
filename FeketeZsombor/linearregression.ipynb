## Data wrangling

### Subtask:
Aggregate the data to count the number of cars per year and state.


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Calculate the average energy consumption
average_energy_consumption = dataset['Energiafogyasztás (kWh)'].mean()

# Reshape data for linear regression
X = dataset['Hőmérséklet (°C)'].values.reshape(-1, 1)
y = dataset['Energiafogyasztás (kWh)'].values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict values for the trend line
y_pred = model.predict(X)


plt.figure(figsize=(10, 6))
plt.scatter(dataset['Hőmérséklet (°C)'], dataset['Energiafogyasztás (kWh)'])
#plt.axhline(y=average_energy_consumption, color='', linestyle='-', label=f'Átlag energiafogyasztás: {average_energy_consumption:.2f} kWh') # Add horizontal line for the average
plt.plot(X, y_pred, color='red', linestyle='-', label='Trendvonal') # Add linear regression trend line
plt.xlabel('Hőmérséklet (°C)')
plt.ylabel('Energiafogyasztás (kWh)')
plt.title('Energiafogyasztás vs Hőmérséklet')
plt.grid(True)
plt.legend() # Add legend to show the label for the average line and trend line
plt.show()
```

**Reasoning**:
Aggregate the data by year and state and count the number of cars in each group.
